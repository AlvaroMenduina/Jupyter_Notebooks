# The Two Body problem and our Solar System

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import time

MU_EARTH = 3.986e14           # Gravitational Parameter: Earth [m^3/s^2]
R_EARTH = 6371000             # Earth's radius [m]
```

## Balance of forces

When a satellite is orbiting over Earth at a distance $r$ to its centre, the only two forces acting on it are the $\textbf{gravitational force}$ $F_g$ and the $\textbf{centrifugal force}$ $F_c$. 

\begin{equation}
F_g = \frac{GMm}{r^2}
\end{equation}

\begin{equation}
F_c = \frac{mV^2}{r}
\end{equation}

Both forces are at equilibrium for the orbit to be circular, so we can write the following:

\begin{equation}
\frac{GMm}{r^2} = \frac{mV^2}{r}
\end{equation}

\begin{equation}
\frac{GM}{r} = V^2
\end{equation}

Which means that the velocity of the satellite in a circular orbit goes as:

\begin{equation}
V = \sqrt{\frac{GM}{r}} = \sqrt{\frac{\mu}{r}}
\end{equation}

This tells us that satellites in a higher orbit (larger $r$) move slowly

```python
h_lower, h_higher = 500e3, 2000e3
r_lower, r_higher = R_EARTH + h_lower, R_EARTH + h_higher

omega_lower = np.sqrt(MU_EARTH / r_lower) / r_lower
omega_higher = np.sqrt(MU_EARTH / r_higher) / r_higher

T_lower = 2*np.pi * np.sqrt(r_lower**3 / MU_EARTH)
T_higher = 2*np.pi * np.sqrt(r_higher**3 / MU_EARTH)

print('Two satellites, one at %d km above the surface (Sat A), the other at %d km (Sat B)' 
      %(int(h_lower/1e3), int(h_higher/1e3)))
print('Sat A completes an orbit in %.1f minutes' %(T_lower/60))
print('Sat B completes an orbit in %.1f minutes' %(T_higher/60))
```

## Our Solar System

```python
MU_SUN = 1.327124e20
# Distances in km
semi_major_axis = {'Mercury': 57909050,
                  'Venus': 108208000,
                  'Earth': 149598023,
                  'Mars': 227939200,
                  'Jupiter': 778.57e6,
                  'Saturn': 1514.50e6,
                   'Uranus': 2875.04e6,
                  'Neptune': 4495.06e6}
```

```python
# True values from Wikipedia
period_days = {'Mercury': 87.969,
              'Venus': 224.701,
               'Earth': 365.256,
              'Mars': 686.971,
              'Jupiter': 4332.59,
              'Saturn': 10759.22,
               'Uranus': 30688.5,
              'Neptune': 60182}
```

```python
def compute_period(a_km, mu=MU_SUN):
    a_m = a_km * 1e3
    T = 2*np.pi * np.sqrt(a_m**3 / mu)
    return T/3600/24
```

```python
planet = 'Venus'
P = compute_period(semi_major_axis[planet])
print('Period of ' + planet + ' %.1f days' %P)
print('Period of ' + planet + ' %.1f years' %(P/365))
```
