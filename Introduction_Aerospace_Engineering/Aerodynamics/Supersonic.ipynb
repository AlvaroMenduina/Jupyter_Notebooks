# Supersonic flight and the sound barrier

We have learned about the different flight regimes according to the definition of the $\textbf{Mach number}$ $M$ and their relation to the local $\textit{sound speed}$. Now let's take a closer look at the details. When an object is flying it emits sounds waves which propagate in all directions. These sound waves travel at the $\textit{speed of sound}$ $c_s$ and look like this:

```python
import numpy as np
import matplotlib.pyplot as plt

# Speed of sound
c_s = 1.
```

```python
def sound_waves(M, n_waves):
    """ 
    Computes the position of sound waves emitted by
    an object travelling at a certain Mach M
    """
    v = c_s * M
    times = np.arange(n_waves)
    t_f = times[-1]
    x_f = v * t_f
    N = 100
    theta = np.linspace(0, 2*np.pi, N)

    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot(111)
    for t in times:
        center = v * t
        radius = (t_f - t) * c_s
        x = center + radius * np.cos(theta)
        y = radius * np.sin(theta)
        ax.plot(x, y, color='blue')

    ax.scatter(x_f, 0.0, color='red', s=70)
    ax.set_aspect('equal')
    plt.title('Sound waves emitted by an object travelling at Mach %.2f' %M)
    plt.axis('off')
    plt.show()
    
    return 
```

```python
sound_waves(M=0.0, n_waves=7)
```

When the object is at rest ($M=0$) all the waves are emitted from the object's position and propagate in a circular fashion. However, when the object is moving ($M>0$) the centre of emission of the sound waves changes as it travels with the object. If the object is moving to the right, the sound waves tend to 'pile up' in that direction

```python
sound_waves(M=0.3, n_waves=7)
```

So the separation between sound waves decreases on one side and increases on the other. This affects the apparent $\textit{frequency}$ of the sound and it's a very famous phenomenon known as $\textbf{Doppler Shift}$.

As an object travels towards us, the apparent frequency of the sound waves grows so that it sounds with a higher pitch. As it moves away from us, the opposite happens and it sounds with a lower pitch

## Subsonic flight ($M<1$)

When an aircraft flies at speeds that are lower than the local $\textit{sound speed}$, the sound waves will always travel ahead of the object, simply because they are travelling faster ($V < c_s$). But you can see from the figure below that at relatively high Mach numbers (around 0.75) the sound waves get piled up quite significantly

```python
sound_waves(M=0.75, n_waves=7)
```

## Sonic flight ($M=1$)

Can you guess what happens when the object is travelling at exactly the $\textit{speed of sound}$?

At $M=1$ the sound waves emitted by the object are all compressed into a $\textit{shock front}$ perpendicular to the direction in which the object is moving. This is what we refer to as a $\textbf{shock wave}$ and it's the cause of the famous sonic boom when an aircraft breaks the sound barrier and enters the supersonic regime.

```python
sound_waves(M=1.0, n_waves=7)

YouTubeVideo('CekJWJFFfHY')
```

## Supersonic flight ($M>1$)

If an aircraft is flying at speed well above the $\textit{speep of sound}$ the sound waves are bound to lag behind it and can only reach a region which is called $\textbf{Mach's cone}$. The semi-angle of the cone ($\alpha$) is directly given by the Mach number:

\begin{equation}
\sin{\alpha} = \frac{c_s}{V} = \frac{1}{M}
\end{equation}

The greater the Mach number, the narrower the Mach cone

```python
sound_waves(M=1.5, n_waves=10)
```
