# Random walk on a Cube

Question:

*Starting at one vertex of a cube, and moving randomly from vertex to adjacent vertices, what is the expected  number of moves until you reach the vertex opposite from your starting point?*

Before mathematically proving the result, let's run some Monte Carlo analysis to simulate this process. We can create a simple Python function that randomly chooses the next vertex from the available neighbours. We will call the starting vertex **0**, and the opposite vertex **X**. The three vertices that are adjacent to **0** will be named as **1, 2, 3** and the vertices adjacent to **X** will be called **A, B, C**

```python
import numpy as np
import matplotlib.pyplot as plt

adjacent_vert = {'0': ['1', '2', '3'],
                 '1': ['0', 'A', 'B'],
                 '2': ['0', 'B', 'C'],
                 '3': ['0', 'A', 'C'],
                 'A': ['X', '1', '3'],
                 'B': ['X', '1', '2'],
                 'C': ['X', '2', '3']}


def cube():
    """
    Starting at vertex 0, randomly move to adjacent vertices
    Until we end up at the opposite vertex X.
    """
    vertex = str(0)
    moves = 0
    while True:
        print("\nAt vertex: ", vertex)
        potential_path = adjacent_vert[vertex]
        choice = np.random.choice(3)
        vertex = potential_path[choice]
        moves += 1
        print("Move to %s | Total moves: %d" % (vertex, moves))
        if vertex == 'X':
            print("Finised")
            break
    return moves
    
```

We can run this cube() function many times and record how many moves were needed to reach the opposite vertex.

```python
N_runs = 1000
paths = []
for k in range(N_runs):
    paths.append(cube())
```

We can show the histogram, and estimate the expected number of moves, as the mean across all runs of the simulation.

```python
print("Expected number of moves to reach the opposite vertex: ", np.mean(paths))


plt.hist(paths, bins=np.arange(3, np.max(paths), 2))
plt.xlabel(r'Number of moves')
plt.ylabel(r'Frequency')
plt.show()
```

## Mathematical proof

The expected number of moves needed to reach the opposite vertex seems to be 10.

Starting at vertex **0**, the expected number of moves can be expressed as:

\begin{equation}
E_0 = 1 + 1/3 \cdot (E_1 + E_2 + E_3)
\end{equation}

the expected number of moves needed to move to the adjacent vertices (**1, 2, 3**), which is always 1, plus the expected number of moves from those vertices. Then, we can also write, the expected number of moves from vertices **1, 2, 3** as:

\begin{equation}
E_1 = 1 + 1/3 \cdot (E_0 + E_A + E_B)
\end{equation}

\begin{equation}
E_2 = 1 + 1/3 \cdot (E_0 + E_B + E_C)
\end{equation}

\begin{equation}
E_3 = 1 + 1/3 \cdot (E_0 + E_A + E_C)
\end{equation}

And the same goes for the vertices adjacent to the opposite vertex **X**:

\begin{equation}
E_A = 1 + 1/3 \cdot (E_1 + E_3 + E_X)
\end{equation}

\begin{equation}
E_B = 1 + 1/3 \cdot (E_1 + E_2 + E_X)
\end{equation}

\begin{equation}
E_C = 1 + 1/3 \cdot (E_2 + E_3 + E_X)
\end{equation}

But we know that $E_X = 0$ because **X** is the final vertex. Also, by the symmetry of the cube, we know that $E_1 = E_2 = E_3 = \alpha$ and $E_A = E_B = E_C = \beta$, so we can simplify the system of equations to:

\begin{equation}
E_0 = 1 + \alpha
\end{equation}

\begin{equation}
\alpha = 1 + 1/3 \cdot (E_0 + 2 \beta)
\end{equation}

\begin{equation}
\beta = 1 + 1/3 \cdot (2 \alpha)
\end{equation}

Substituting the first and third equation into the second:

\begin{equation}
\alpha = 1 + 1/3 \cdot (3  + 7/3 \alpha)
\end{equation}

\begin{equation}
3\alpha = 6  + 7/3 \alpha
\end{equation}

we get:

\begin{equation}
\alpha = 9 = E_1
\end{equation}

\begin{equation}
E_0 = 10
\end{equation}

\begin{equation}
\beta = 7
\end{equation}

So the expected number of moves to go from one vertex to the opposite is 10, as suggested by our Monte Carlo analysis. A surprising part of this result is that the number of moves if starting from vertices **A, B, C** is 7, which seems rather large considering those vertices are already adjacent to the final **X**, but there's a 2/3 chance of moving to **1, 2, 3**, whose expected number of moves is 9.


## Another perspective

Another way to think about this problem is in terms of *mistakes*. The minimum number of moves required to reach the opposite vertex is 3. The probability of this happening is: $1 \cdot 2/3 \cdot 1/3 = 2/9$ since the first move doesn't matter, and at the second move we have 2/3 chance of choosing the right vertex, and at the third move we have only 1/3 chance of choosing **X**.

The next possibility is making one mistake by going back to an incorrect vertex, which will lead to a total of 5 moves (3 + 1 mistake + 1 move to fix it). The probability of doing it in 5 steps can be expressed as the probability of making one mistake (7/9) times the probability of doing everything else right (2/9)

This can be extended to an arbitrary case of making $n$ mistakes: $P(n) = (7/9)^n (2/9)$ and finishing after $3 + 2n$ moves. Using the expression for the Expectation:

\begin{equation}
E[X] = \sum_n^{\infty} X_n P(n) = \sum_n^{\infty} (3 + 2n)(7/9)^n (2/9)
\end{equation}

which gives the same result as $\sum_n^{\infty} (3 + 2n)(7/9)^n = 45$, and $45 \cdot 2/9 = 10$


## A Markov chain viewpoint

We can regard this problem as a Markov chain with 4 possible states $S$, defined by the distance to the opposite vertex: $S = {3, 2, 1, 0}$. For example, the starting point is 3 moves away so we'll call it $S_3$. Based on this we can construct the *transition matrix* $P$ with entries $P_{ij}$ defined as the probability of moving from state $S_i$ to $S_j$ in a single step.

\begin{equation}
P = 
\begin{bmatrix}
P(S_3 \rightarrow S_3) & P(S_3 \rightarrow S_2) & P(S_3 \rightarrow S_1) & P(S_3 \rightarrow S_0) \\
P(S_2 \rightarrow S_3) & P(S_2 \rightarrow S_2) & P(S_2 \rightarrow S_1) & P(S_2 \rightarrow S_0) \\
P(S_1 \rightarrow S_3) & P(S_1 \rightarrow S_2) & P(S_1 \rightarrow S_1) & P(S_1 \rightarrow S_0) \\
P(S_0 \rightarrow S_3) & P(S_0 \rightarrow S_2) & P(S_0 \rightarrow S_1) & P(S_0 \rightarrow S_0) \\
\end{bmatrix}
\end{equation}

\begin{equation}
P = 
\begin{bmatrix}
0 & 1 & 0 & 0 \\
1/3 & 0 & 2/3 & 0 \\
0 & 2/3 & 0 & 1/3 \\
0 & 0 & 0 & 1 \\
\end{bmatrix}
\end{equation}

For the beginning state $S_3$, we always move to a state $S_2$ with probability 1. At state $S_2$, we have 1/3 chance of going back to $S_3$ and 2/3 chance of progressing to $S_1$. For state $S_1$, 2/3 chance of going back to $S_2$ and 1/3 chance of moving to $S_0$. Finally, since $S_0$ is our goal we stay there with probability 1.

State $S_0$ is what we can in Markov chain terminology an **absorbing state**, since once it is reached we cannot leave it. From Markov chain theory, the transition matrix $P$ for $t$ transitioning states and $r$ absorbing states can be divided into:

\begin{equation}
P = 
\begin{bmatrix}
Q_t & R \\
0_{r, t} & I_r
\end{bmatrix}
\end{equation}

where $Q_t$ is a $t \times t$ matrix that represents the probability of jumping from one transitioning state to another transitioning state, $R$ is a $t \times r$ matrix representing the probability of jumping from a transitioning state to an absorbing state, $0_{r, t}$ represents the fact that one cannot jump from an absorbing state to a transitioning state, and $I_r$ is the identity matrix representing the number of absorbing states.

With this we can define the matrix $(I_t - Q_t)^{-1}$ whose entries $ij$ represent the expected number of visits to state $j$ until being absorbed starting at state $i$.

\begin{equation}
(I_t - Q_t)^{-1} = 
\begin{bmatrix}
2.5 & 4.5 & 3 \\
1.5 & 4.5 & 3 \\
1 & 3 & 3 \\
\end{bmatrix}
\end{equation}

The first row indicates that, starting from state $S_3$ we visit $S_3$ 2.5 times on average, $S_2$ 4.5 times and $S_1$ 3 times, before being absorbed at $S_0$. **The expected number of steps to get from $S_3$ to $S_0$ is thus given by the sum over all state visits in that row, which is 10. ** The sum over the second row represents that expected number of steps if starting from $S_2$, which is 9 as already shown in the previous section.


### One step further

Another interesting link to the previous section is that the two-step transition matrix, i.e. the probability of the system being in state $j$ after 2 steps starting at state $i$ is given by the square of $P$. If we calculate this we see that:

\begin{equation}
P^2 = 
\begin{bmatrix}
1/3 & 0 & 2/3 & 0 \\
0 & 7/9 & 0 & 2/9 \\
2/9 & 0 & 4/9 & 1/3 \\
0 & 0 & 0 & 1 \\
\end{bmatrix}
\end{equation}

Starting at $S_3$ we have 1/3 chance of being back at $S_3$ after two steps, and 2/3 chance of being at state $S_1$. The last entry of first row represents the probability of being at $S_0$ after 2 moves from $S_3$, which is obviously 0. If we repeat that for $P^3$, we find:

\begin{equation}
P^3 = 
\begin{bmatrix}
0 & 0 & 7/9 & 2/9 \\
7/27 & 0 & 14/27 & 2/9 \\
0 & 14/27 & 0 & 13/27 \\
0 & 0 & 0 & 1 \\
\end{bmatrix}
\end{equation}

the chance of being at state $S_0$ after 3 steps starting at $S_3$ is 2/9, just like we showed in the previous section. This can extended to define the expectation for the number of moves as a sum over the powers of the state transition matrix $P$.

We can use this idea to ask ourselves another question about this problem. 

*We place 10 ants at every vertex of the cube, except for the final vertex **X** and let them follow go from one vertex to another randomly. If they reach **X**, they stay there. What is the expected number of ants at **X** after 10 moves* 

We can define the initial state vector as $S^0$, the number of ants in each state (remember the definition of state is given by the distance to the final vertex) as $S^0 = [10, 30, 30, 0]$. Using Markov chain properties, the state vector after 1 move $S^1$ is given by the product of the state vector $S^0$ and the transition matrix $P$: $S^1 = S^0 \cdot P$

```python
P = np.array([[0, 1, 0, 0], 
            [1/3, 0, 2/3, 0],
             [0, 2/3, 0, 1/3],
             [0, 0, 0, 1]])

s_0 = np.array([10, 30, 30, 0])

s_1 = np.dot(s_0.T, P)
s_1
```

After 1 step, on average, 10 ants will be at $S_3$ because there were 30 ants in $S_2$ and 1/3 will have moved back. The 10 ants initially at $S_3$ will have moved to $S_2$, alongside 2/3 of the ants initially at $S_1$, for a total of 30 ants. There will be 20 ants at $S_1$ who have come from $S_2$ (2/3) and there will be 10 ants from $S_1$ that succesfully made it to $S_0$.

After **2 steps**, the results will be given by $S^2 = S^1 \cdot P = S^0 \cdot P^2$, where $P^2$ indicates the power 2 for the transition matrix.

```python
s_2 = np.dot(s_0.T, np.linalg.matrix_power(P, 2))
s_2
```

And for **10 steps**, we find that around 50 out of the 70 ants will have made it to **X**.

```python
s_10 = np.dot(s_0.T, np.linalg.matrix_power(P, 10))
s_10
```

In a sense, what this dot product is calculating is the *expected* number of ants at each vertex, where a given result is the product of the original distribution of ants and the probabilities of ending at that given vertex after 10 steps:

\begin{equation}
S^{10}(S_3) = S^3 P^{10}(S_3 \rightarrow S_3) + S^2 P^{10}(S_2 \rightarrow S_3) + S^1 P^{10}(S_1 \rightarrow S_3) + S^0 P^{10}(S_1 \rightarrow S_3)
\end{equation}

We can use these ideas to calculate how the expected population of ants at vertex **X** evolves with the number of steps

```python
MAX = 50
n_steps = range(MAX + 1)
x_ants = []
for n in n_steps:
    s_n = np.dot(s_0.T, np.linalg.matrix_power(P, n))
    x_ants.append(s_n[-1] / np.sum(s_0))

fig, ax = plt.subplots(1, 1)
ax.plot(n_steps, x_ants)
ax.set_xlabel(r'Number of steps $n$')
ax.set_ylabel(r'Ratio of ants at vertex X')
ax.set_ylim([0, 1])
ax.set_xlim([0, MAX])
plt.show()
```
