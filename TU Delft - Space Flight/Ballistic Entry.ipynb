# Re-Entry Systems 
## [1] Ballistic Entry

![title](img/apollo.jpeg)

*Ballistic entry* refers to a flight where no lift is acting on the vehicle, and its atmospheric trajectory is purely governed by drag and gravity. This is typical of Mars entry capsules, sample return vehicles and entry probes like Huygens (entry and descend through Titan's atmosphere). To a certain extent, this is also applicable to the Apollo return capsule, at least during a portion of its trajectory since, in order to reduce loads and ensure some maneuvrability, it flew with a slight angle of attack to generate some lift.

In this notebook, we will derive the equations of motion for this type of trajectory. We will investigate the altitude-velocity profile as well as the maximum deceleration, looking at the impact of the different parameters and initial conditions.

**Equations of motion**

Starting with the general equations of motion for the vehicle:

\begin{equation}
m \frac{dV}{dt} = -D - mg \sin{\gamma}
\end{equation}

\begin{equation}
m V\frac{d\gamma}{dt} = L - mg \cos{\gamma}\left(1 - \frac{V^2}{g r}\right)
\end{equation}


\begin{equation}
\frac{dr}{dt} = \frac{dh}{dt} = V \sin{\gamma}
\end{equation}

## Ballistic Entry: Flight Path

Under the assumption of "No Lift", the equation for the perpendicular direction can be expressed as:

\begin{equation}
V\frac{d\gamma}{dt} = - g \cos{\gamma}\left(1 - \frac{V^2}{g r}\right)
\end{equation}

Using the fact that:

\begin{equation}
\frac{d\gamma}{dt} = \frac{d\gamma}{dh} \frac{dh}{dt} = \frac{d\gamma}{dh} V \sin{\gamma}
\end{equation}

we can write the equation in terms of height $h$ rather than as a function of time $t$:

\begin{equation}
V^2\sin{\gamma} \frac{d\gamma}{dh} = - g \cos{\gamma}\left(1 - \frac{V^2}{V^2_c}\right)
\end{equation}

rearranging:

\begin{equation}
\tan{\gamma} \frac{d\gamma}{dh} = - g \left(\frac{1}{V^2} - \frac{1}{V^2_c}\right) \simeq 0
\end{equation}

since the entry velocity $V$ is close to the local circular orbital velocity $V_C$, we find that the change in flight path angle $\gamma$ is approximately 0, meaning that it remains almost constant during flight $\gamma \simeq \gamma_E$. This is specially true for steep entries where $\gamma$ is relatively large, and $\tan{\gamma}$ further reduces the derivative. Therefore, the **ballistic entry** trajectory is approximately a straight line, defined by the initial entry angle $\gamma_E$.

For steep angles $\gamma$ and given the thickness of the atmosphere $h\sim100$ km, it is not necessary to consider the Earth's curvature when calculating the travelled distance.

## Ballistic Entry: Altitude - Velocity profile

An important thing to consider for a ballistic entry is the *altitude-velocity* profile, or how the velocity evolves as the vehicle travels through the atmosphere. Using the fact that:


\begin{equation}
\frac{dV}{dt} = \frac{dV}{dh} \frac{dh}{dt} = V \sin{\gamma}\frac{dV}{dh} 
\end{equation}

we can plug that into the *drag* equation, and dividing by $m g \sin{\gamma}$, we obtain:

\begin{equation}
\frac{V}{g} \frac{dV}{dh} = - \frac{D}{m g \sin{\gamma}} - 1
\end{equation}

since the drag is large, the first term dominates over 1, and we can simplify to:

\begin{equation}
\frac{V}{g} \frac{dV}{dh} \simeq - \frac{\frac{1}{2}\rho V^2 S C_D}{m g \sin{\gamma}} = - \frac{\rho V^2}{2 K \sin{\gamma}}
\end{equation}

where $K=\frac{m g}{S C_D}$ is the **ballistic parameter**.
Rearrainging:

\begin{equation}
\frac{dV}{dh} = - \frac{1}{2}\frac{g \rho dh}{K\sin{\gamma}} = \frac{1}{2}\frac{dp}{K\sin{\gamma}}
\end{equation}

solving for $V$ and using the formula for the standard atmosphere $p = \rho R T = \rho_0 g /\beta\exp{(-\beta h)}$:

\begin{equation}
\ln{\frac{V}{V_E}} = \frac{1}{2}\frac{p - p_E}{K\sin{\gamma}}
\end{equation}


\begin{equation}
\frac{V}{V_E} = \exp{\left(\frac{1}{2}\frac{g \rho_0 \exp{(-\beta h)}}{K \beta\sin{\gamma_E}} \right)}
\end{equation}

Which tells us that the ratio between the velocity at a given height $V(h)$ and the initial entry velocity $V_E$ depends on $h$ and other parameters such as the **ballistic parameter** $K$, the initial flight path angle $\gamma_E$, and the characteristics of the atmosphere $\rho_0$, $\beta$.

This allows us to investigate the impact of the different parameters on the altitude-velocity profile. In the code below we look at how the **ballistic parameter** $K$ and the entry angle $\gamma_E$ affect the profile.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

plt.rc('font', family='serif')
plt.rc('text', usetex=False)

# Parameters
g = 9.81            # gravity [m / s^2]
rho0 = 1.225        # density at sea level [kg / m^3]
beta = 1 / 7050     # atmosphere height scale factor [1 / m]
```

```python
# We look at a height profile from 100 km to sea-level
h = 1e3 * np.linspace(100, 0, 100)

gammas = [5, 10]   # several values of entry angle [deg]
K_ballistic = [200, 2000, 20000, 500000]      # ballistic coefficient [N / m^2]
colors = cm.Reds(np.linspace(0.5, 1.0, len(K_ballistic)))
linestyles = ['-', '--']

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
for i, K in enumerate(K_ballistic):

    for j, gamma in enumerate(gammas):
        gamma_entry = -np.deg2rad(gamma)
        A = 0.5 * g * rho0 * np.exp(-h * beta)
        B = K * beta * np.sin(gamma_entry)
        v_ratio = np.exp(A / B)
        ax.plot(v_ratio, h / 1000, color=colors[i], linestyle=linestyles[j], 
                label="K=%d, $\gamma=-%d$ deg" % (K, gamma))

ax.set_xlim([0, 1])
ax.set_ylim([0, 100])
ax.legend()
ax.grid(True)
ax.set_xlabel(r'Velocity ratio $\frac{V}{V_E}$')
ax.set_ylabel(r'Height [km]')
ax.set_title(r'Ballistic Entry : Altitude - Velocity profile')
plt.show()
```

The flight path begins at the top-right corner, where the height is 100 km (entrance of the atmosphere) when the velocity is equal to the entry velocity $V_E$.
The entry vehicle decays in height very quickly at the beginning given the initial velocity, and then the atmospheric drag starts to decelerate it.

Looking at the **ballistic parameter** we can notice the following effect. As $K$ increases, meaning either a heavier vehicle ($m g$) or one capable of generating less drag $C_D S$, the vehicle penetrates deeper into the atmosphere before the drag manages to decelerate it sufficiently. For a given terminal velocity, this will be reached at lower altitudes if $K$ is larger.

Interestingly, if the vehicle is too heavy or if it's not capable of generating enough drag, it will reach the surface before it has sufficiently decelerated, crashing at great speed. Therefore, this altitude-velocity profile can be used to calculate the vehicle characteristics needed for the entry to be feasible. For example, it can be used to calculate a bound on how heavy the capsule can be, in order to safely enter a planet's atmosphere given the aerodynamic design of the heat shield.

With regards to the **flight path angle** $\gamma$ the effect is similar to the ballistic parameter. Steeper entry angles make the vehicle reach lower altitudes before atmospheric drag manages to slow it down.

## Ballistic Entry: Deceleration

One of the most important considerations for a re-entry vehicle is its maximum deceleration, as this determines the maximum mechanical load it needs to withstand during re-entry; and it's specially important for manned vehicles where we want to keep the accelerations to a minimum. In order to calculate the deceleration, we start with:


\begin{equation}
m \frac{dV}{dt} \simeq -D
\end{equation}

so that:

\begin{equation}
a = -\frac{dV}{dt} = \frac{\frac{1}{2} \rho V^2 C_D S g}{m g} = \frac{g}{2K}\rho V^2
\end{equation}

using the velocity equation from the previous section, we can obtain a formula for $\rho$:

\begin{equation}
\rho = \frac{2 K \beta \sin{\gamma_E}}{g} \ln{\frac{V}{V_E}}
\end{equation}

that we can plug back into the acceleration equation:

\begin{equation}
a = -\frac{dV}{dt} = \beta \sin{\gamma_E} V^2_E \left(\frac{V}{V_E}\right)^2 \ln{\frac{V}{V_E}}
\end{equation}

This can be used to calculate the *maximum* deceleration with $\frac{d a}{d(V / V_E)} = 0$, which gives rise to:

\begin{equation}
\frac{V'}{V_E} = \frac{1}{\sqrt{e}} = 0.606
\end{equation}

so that the *maximum* deceleration occurs when the velocity is about 60% of the entry velocity. The value for the *maximum* deceleration is given by:

\begin{equation}
a_{max} = - \frac{\beta \sin{\gamma_E}}{2e} V^2_E
\end{equation}

We can also calculate the height at which that *maximum* deceleration happens:

\begin{equation}
h(a_{max}) = \frac{1}{\beta} \ln{ \left(-\frac{\rho_0 g}{K \beta \sin{\gamma_E}} \right)}
\end{equation}



```python
# We look at a height profile from 100 km to sea-level
h = 1e3 * np.linspace(100, 0, 200)
gammas = [5, 10]   # several values of entry angle [deg]
K_ballistic = [200, 2000, 20000]      # ballistic coefficient [N / m^2]
colors = cm.Reds(np.linspace(0.5, 1.0, len(K_ballistic)))
linestyles = ['-', '--']

# Assuming an entry
v_e = 7e3     # 7 km / s

fig, axes = plt.subplots(1, 2, figsize=(20, 10))
for i, K in enumerate(K_ballistic):
    for j, gamma in enumerate(gammas):
        gamma_entry = -np.deg2rad(gamma)

        A = 0.5 * g * rho0 * np.exp(-h * beta)
        B = K * beta * np.sin(gamma_entry)
        v_ratio = np.exp(A / B)

        a = beta * np.sin(gamma_entry) * v_e**2 * v_ratio**2 * np.log(v_ratio) / g
        ax = axes[j]
        ax.plot(a, h / 1000, color=colors[i], 
                    label="K=%d, $\gamma=-%d$ deg" % (K, gamma))

for ax in axes:
    ax.set_xlim([0, 30])
    ax.set_ylim([0, 100])
    ax.legend()
    ax.grid(True)
    ax.set_xlabel(r'Deceleration $\frac{a}{g}$')
    ax.set_ylabel(r'Height [km]')
    ax.set_title(r'Ballistic Entry : Deceleration profile')
plt.show()
```

Looking at the two equations and the deceleration profiles from the figure above, we can draw the following conclusions. 

The first thing to note is that the *maximum* deceleration does **not** depend on the **ballistic parameter**. For a given flight profile, all vehicles experience the same *maximum* deceleration independent of their mass and drag characteristics. But the altitude at which this deceleration happens varies. As we saw for the altitude-velocity profile, heavier vehicles suffer the maximum deceleration deeper into the atmosphere.

The second thing to note is that the height at which the *maximum* deceleration takes place does **not** depend on the initial velocity $V_E$. In contrast, both the *maximum* deceleration and height depend on the initial entry angle $\gamma_E$. Steeper flight paths lead to larger decelerations, suggesting the need to properly control the angle $\gamma$ to minimise mechanical loads. Steeper paths also move the height of maximum deceleration deeper into the lower atmosphere.

## Thermal Loads

Another crucial element of the ballistic trajectory are the thermal loads that the vehicle has to withstand during deceleration. In order to model these, we can make use of the Chapman equation which tells us that the heat flux due to convection at the stagnation point is given by:

\begin{equation}
q = c_1 \frac{1}{R_N^n}\rho^{1 - n} V^m
\end{equation}

where $c_1$ is a constant, $R_N$ is the nose radius for the vehicle, $n$ is an exponent that depends on the boundary layer conditions ($n=0.5$ for laminar flow, $n=0.2$ for turbulent), and $m=3$ is an empirical exponent. This explains why entry vehicles are designed with a blunt nose of large nose radius $R_N$ that minimises the heat loads.

Using the expression for the density as a function of the velocity ratio, we can calculate the dependency of the heat flux $q$ on the velocity as:

\begin{equation}
q \propto \left|\ln{\frac{V}{V_E}} \right|^{1-n} \left(\frac{V}{V_E} \right)^3
\end{equation}

We can use this to calculate the **maximum** heat flux, differentiating with respect to the velocity ratio, so that we find the maximum at:

\begin{equation}
\frac{V''}{V_E} = \exp{\left(- \frac{1 - n}{3} \right)}
\end{equation}

Depending on the flow conditions, the **maximum** heat flux occurs at 0.85 or 0.77 of velocity ratio. If we compare this to the deceleration load from the previous section, we see that a vehicle following a ballistic entry will experience the condition for **maximum** heat flux first and at a higher altitude, followed by the *maximum* deceleration at a lower altitude.

Comparing both expressions for the deceleration and heat flux, we can find a formula for the height difference as:

\begin{equation}
h'' - h' = - \frac{\ln{2/3}(1 - n)}{\beta}
\end{equation}

which tells us that the condition for **maximum** heat flux usually occurs between 4.5 and 7.9 km above the *maximum* deceleration point. 

Interestingly, this altitude difference does not depend on any parameter that we can control such as the entry angle $\gamma$ or the ballistic parameter $K$. One might want to separate the two conditions of **maximum** heat flux (thermal loads) and *maximum* deceleration (mechanical load) because the aero-heating weakens the structure of the vehicle against mechanical loads. However, this altitude difference $h'' - h'$ is fixed to just a few km. Since the vehicle is travelling at approximately the local circular velocity ($\simeq 7$ km / s) this implies that the *maximum* deceleration happens only around 1 second after the vehicle has received the **maximum** heatflux.

```python
h = 1e3 * np.linspace(100, 0, 200)
gamma = 10
K = 2000

gamma_entry = -np.deg2rad(gamma)

A = 0.5 * g * rho0 * np.exp(-h * beta)
B = K * beta * np.sin(gamma_entry)
v_ratio = np.exp(A / B)

# a / a_max
a_ratio = - 2 * np.exp(1) * v_ratio**2 * np.log(v_ratio)

# q / q_max
n = 0.5
q_ratio = (3 * np.exp(1) / (1 - n)) ** (1 - n) * np.abs(np.log(v_ratio)) ** (1 - n) * v_ratio**3

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.plot(v_ratio, a_ratio, label='Deceleration')
ax.plot(v_ratio, q_ratio, label=r'Heat Flux ($n=%.1f$)' % n)
ax.set_xlabel(r'Velocity ratio $\frac{V}{V_E}$')
ax.set_xticks(np.arange(0, 1.1, 0.1))
ax.set_ylabel(r'Ratio to max [ ]')
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.grid(True)
ax.legend(loc=2)
plt.show()
```

## Ballistic Entry : Maximum Deceleration profile

Another interesting thing we can do is to calculate the *maximum* deceleration and the altitude at which that happens as a function of the entry angle $\gamma$ and the characteristics of the vehicle, in the form of the ballistic parameter $K$. 

This can be used to select adequate mission parameters given a set of constraints such as the maximum acceptable deceleration (g-load), or the height at which this can take place.

```python
# We look at a height profile from 100 km to sea-level
h = 1e3 * np.linspace(100, 0, 500)
K_ballistic = [200, 2000, 20000]      # ballistic coefficient [N / m^2]
colors = cm.Reds(np.linspace(0.5, 1.0, len(K_ballistic)))
linestyles = ['-', '--']

# Assuming an entry
v_e = 7e3     # 7 km / s

Ng = 9
gammas = np.linspace(5, 15, Ng)

fig, ax = plt.subplots(1, 1, figsize=(8, 8))
for i, K in enumerate(K_ballistic):
    h_list, a_list = [], []
    for j, gamma in enumerate(gammas):
        gamma_entry = -np.deg2rad(gamma)

        A = 0.5 * g * rho0 * np.exp(-h * beta)
        B = K * beta * np.sin(gamma_entry)
        v_ratio = np.exp(A / B)

        a = beta * np.sin(gamma_entry) * v_e**2 * v_ratio**2 * np.log(v_ratio) / g
        
        # calculate the maximum acceleration
        k_max = np.nanargmax(a)
        a_max = np.nanmax(a)
        h_amax = h[k_max] / 1000
        h_list.append(h_amax)
        a_list.append(a_max)
    color = colors[i]
    ax.plot(a_list, h_list, color=color, label="K=%d" % (K))
    if i == 0:
        ax.scatter(a_list[0], h_list[0], label=r'$\gamma=%.1f$ deg' % gammas[0], marker='^', color=color)
        ax.scatter(a_list[Ng//2], h_list[Ng//2], label=r'$\gamma=%.1f$ deg' % gammas[Ng//2], marker='o', color=color)
        ax.scatter(a_list[-1], h_list[-1], label=r'$\gamma=%.1f$ deg' % gammas[-1], marker='v', color=color)
    else:
        ax.scatter(a_list[0], h_list[0], marker='^', color=color)
        ax.scatter(a_list[Ng//2], h_list[Ng//2], marker='o', color=color)
        ax.scatter(a_list[-1], h_list[-1], marker='v', color=color)

ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.legend()
ax.grid(True)
ax.set_xlabel(r'Deceleration $\frac{a}{g}$')
ax.set_ylabel(r'Height [km]')
ax.set_title(r'Ballistic Entry : Maximum Deceleration')
plt.show()
```
